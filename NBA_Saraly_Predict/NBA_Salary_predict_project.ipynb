# NBA salary predict

## About the Dataset

This dataset merges player per-game and advanced statistics for the NBA's 2022-23 season with player salary data, creating a comprehensive resource for understanding the performance and financial aspects of professional basketball players. The dataset is the result of web scraping player salary information from Hoopshype, and downloading traditional per-game and advanced statistics from Basketball Reference.

### Source: https://www.kaggle.com/datasets/jamiewelsh2/nba-player-salaries-2022-23-season/data

* Player Information: Player name, team(s) played for during the season.
* Salary: This column typically represents the salary or earnings of each player. It is the amount of money a player is paid for their services in the NBA.
* Position: The "Position" column indicates the primary playing position of each player on the basketball court, such as Point Guard (PG), Shooting Guard (SG), Small Forward (SF), Power Forward (PF), or Center (C).

* Age: This column represents the chronological age of the player. It is measured in years and is an important factor in assessing a player's experience and physical condition.

* Team: The "Team" column identifies the NBA team that the player is currently or was associated with during the season in question.

* GP (Games Played): "GP" indicates the number of games in which the player participated during the season.

* GS (Games Started): "GS" represents the number of games in which the player was in the starting lineup.

* MP (Minutes Played): This column records the total number of minutes the player was on the court during games.

* FG (Field Goals Made): It indicates the total number of successful field goals (baskets) made by the player.

* FGA (Field Goals Attempted): "FGA" is the total number of field goal attempts made by the player, regardless of success.

* FG% (Field Goal Percentage): It represents the shooting accuracy, calculated as the percentage of successful field goals made out of the total attempts.

* 3P (Three-Pointers Made): This column indicates the total number of successful three-point shots made by the player.

* 3PA (Three-Point Attempts): "3PA" represents the total number of three-point shot attempts by the player.

* 3P% (Three-Point Percentage): It is the shooting accuracy for three-pointers, calculated as the percentage of successful three-point shots made out of total attempts.

* 2P (Two-Pointers Made): "2P" records the total number of successful two-point field goals made by the player.

* 2PA (Two-Point Attempts): "2PA" is the total number of two-point field goal attempts.

* 2P% (Two-Point Percentage): It represents the shooting accuracy for two-point field goals, calculated as the percentage of successful two-pointers made out of total attempts.

* eFG% (Effective Field Goal Percentage): eFG% is a more comprehensive measure of field goal accuracy that accounts for the extra value of three-pointers.

* FT (Free Throws Made): "FT" indicates the total number of successful free throws made by the player.

* FTA (Free Throw Attempts): "FTA" represents the total number of free throw attempts.

* FT% (Free Throw Percentage): It is the free throw shooting accuracy, calculated as the percentage of successful free throws made out of total attempts.

* ORB (Offensive Rebounds): "ORB" records the number of offensive rebounds grabbed by the player.

* DRB (Defensive Rebounds): "DRB" represents the number of defensive rebounds grabbed by the player.

* TRB (Total Rebounds): "TRB" is the total number of rebounds (both offensive and defensive) grabbed by the player.

* AST (Assists): This column indicates the total number of assists made by the player.

* STL (Steals): "STL" represents the total number of steals made by the player.

* BLK (Blocks): "BLK" records the total number of blocks (shots blocked) made by the player.

* TOV (Turnovers): This column indicates the total number of turnovers committed by the player.

* PF (Personal Fouls): "PF" represents the total number of personal fouls committed by the player.

* PTS (Points): "PTS" records the total number of points scored by the player during the season.

* Total Minutes: This column represents the sum of minutes played by the player throughout the season.

* PER (Player Efficiency Rating): PER is a player performance metric that attempts to measure a player's overall contribution to their team.

* TS% (True Shooting Percentage): TS% is a measure of shooting efficiency that takes into account field goals, three-pointers, and free throws.

* 3PAr (Three-Point Attempt Rate): 3PAr is the percentage of field goal attempts that are three-pointers.

* FTr (Free Throw Attempt Rate): FTr is the percentage of field goal attempts that result in free throw attempts.

* ORB% (Offensive Rebound Percentage): ORB% is the percentage of available offensive rebounds grabbed by the player.

* DRB% (Defensive Rebound Percentage): DRB% is the percentage of available defensive rebounds grabbed by

* TRB% (Total Rebound Percentage): TRB% represents the percentage of available rebounds (both offensive and defensive) that a player grabs while they are on the court. It is a measure of a player's effectiveness in securing rebounds.

* AST% (Assist Percentage): AST% is the percentage of a team's made field goals for which a player is credited with an assist. It quantifies a player's ability to create scoring opportunities for teammates.

* STL% (Steal Percentage): STL% is the percentage of opponent possessions that end with a steal by the player. It measures a player's effectiveness in stealing the ball from opponents.

* BLK% (Block Percentage): BLK% is the percentage of opponent two-point field goal attempts blocked by the player. It assesses a player's shot-blocking abilities.

* TOV% (Turnover Percentage): TOV% represents the percentage of a player's possessions that result in a turnover. It quantifies a player's ball-handling and decision-making efficiency.

* USG% (Usage Percentage): USG% is an estimate of the percentage of team plays used by a player while they are on the court. It reflects how involved a player is in their team's offense.

* OWS (Offensive Win Shares): OWS is an estimate of the number of wins contributed by a player's offensive performance.

* DWS (Defensive Win Shares): DWS is an estimate of the number of wins contributed by a player's defensive performance.

* WS (Win Shares): WS is the sum of Offensive Win Shares and Defensive Win Shares. It is an overall metric that assesses a player's total contribution to winning games.

* WS/48 (Win Shares per 48 Minutes): WS/48 is a rate statistic that estimates the number of win shares a player would contribute per 48 minutes of play.

* OBPM (Offensive Box Plus-Minus): OBPM is a box score-based metric that estimates a player's offensive contributions in terms of points per 100 possessions above the league average.

* DBPM (Defensive Box Plus-Minus): DBPM is a box score-based metric that estimates a player's defensive contributions in terms of points per 100 possessions above the league average.

* BPM (Box Plus-Minus): BPM is the sum of Offensive Box Plus-Minus and Defensive Box Plus-Minus. It provides an overall assessment of a player's contributions on both ends of the court.

* VORP (Value Over Replacement Player): VORP is a box score-based metric that quantifies a player's value compared to a hypothetical replacement-level player. It estimates the player's contribution to team success.

## Train Test Split

```python
# set variables for train test split
X = df_clean_1.drop(['Salary'], axis = 1)
y = df_clean_1["Salary"]
```

```python
# Divide data for train test split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = .30, random_state=0)
```

```python
# Show Data in table X_train
len(X_train)
```

```python
# Show Data in table X_test
len(X_test)
```

## Linear Regression Model

```python
# Create a Linear Regression model
model_lin = LinearRegression()

# Train the model on the training data
model_lin.fit(X_train, y_train)

# Make predictions on the test data
y_pred_lin = model_lin.predict(X_test)

# Evaluate model
# Calculate MAE
mae_lin = mean_absolute_error(y_test, y_pred_lin)

# Calculate MSE
mse_lin = mean_squared_error(y_test, y_pred_lin)

# Calculate RMSE
rmse_lin = np.sqrt(mean_squared_error(y_test, y_pred_lin))

# Calculate R2
r2_lin = r2_score(y_test, y_pred_lin)
```

```python
# Show MAE, MSE, RMSE, R2
print("Mean Absolute Error (MAE):", mae_lin)
print("Mean Squared Error (MSE):", mse_lin)
print("ROOT Mean Squared Error (RMSE):", rmse_lin)
print("R2_SCORE (R2):", r2_lin)
```

```python
# Plot actual vs. predicted values for LinearRegression Model
fig, ax = plt.subplots(figsize=(20, 6))
y_test_array = np.array(y_test)
plt.plot(y_test_array[:], marker='o', label='y_test')
plt.plot(y_pred_lin[:], color='red', marker='o', label='predict')
plt.title('y_test vs predict', size=25, y=1.01)
plt.legend()

# Display the plot
plt.show()
```

# Model Comparison
The less the Root Mean Squared Error (RMSE), The better the model is.

```python
model = ["Linear Regression", "Random Forest Regression", "Gradient Boosting Regression", "Neural Networks", "XGBoost", "Lasso Regression" ]
mae = [mae_lin, mae_rf, mae_reg, mae_clf, mae_xgb, mae_lasso]
mse = [mse_lin, mse_rf, mse_reg, mse_clf, mse_xgb, mse_lasso]
rmse = [rmse_lin, rmse_rf, rmse_reg, rmse_clf, rmse_xgb, rmse_lasso]
r2 = [r2_lin, r2_rf, r2_reg, r2_clf, r2_xgb, r2_lasso]

df_mae_mse = pd.DataFrame({
    "Model": model,
    "MAE": mae,
    "MSE": mse,
    "RMSE": rmse,
    "r2" : r2
})
df_mae_mse
```

```python
# Sort the DataFrame by RMSE values in ascending order
df_mae_mse_sorted = df_mae_mse.sort_values(by='RMSE')

# Create a bar plot for RMSE with sorted data
plt.figure(figsize=(10, 6))
plt.barh(df_mae_mse_sorted["Model"], df_mae_mse_sorted["RMSE"], color='skyblue')
plt.xlabel('RMSE', fontsize=12)
plt.title('Root Mean Squared Error (RMSE) for Different Models (Sorted)', fontsize=14)
plt.gca().invert_yaxis()  # Invert the y-axis to display the lowest RMSE at the top
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Display the sorted plot
plt.show()
```
